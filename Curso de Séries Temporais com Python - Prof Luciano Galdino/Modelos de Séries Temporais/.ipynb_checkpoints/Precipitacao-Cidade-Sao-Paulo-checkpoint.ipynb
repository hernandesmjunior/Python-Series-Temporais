# 0.0 Importações

## 0.1 Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.tsa.stattools
import warnings

from matplotlib.pylab import rcParams
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
from statsmodels.tsa.arima_model import ARIMA
from pmdarima.arima import auto_arima
from sklearn.metrics import mean_absolute_error, mean_squared_error


sns.set()
```

```python
warnings.filterwarnings('ignore')
```

## 0.2 Dataset

```python
chuva = pd.read_csv('Data/E3-262_Chuva_Mensal.csv', sep=';')
```

```python
chuva.head()
```

```python
chuva.shape
```

# 1.0 Análise e Tratamento Inicial

```python
# verificando a tipagem

chuva.dtypes
```

```python
# substituir as vírgulas por pontos

chuva = chuva.apply(lambda x: x.str.replace(',','.'))
```

```python
# tranformar variáveis categórias em numéricas e substituir os traços pela média

chuva['Janeiro'] = pd.to_numeric(chuva['Janeiro'].str.replace('---', '305.79'))

chuva['Fevereiro'] = pd.to_numeric(chuva['Fevereiro'].str.replace('---', '249.3'))

chuva['Março'] = pd.to_numeric(chuva['Março'].str.replace('---', '227.97'))

chuva['Abril'] = pd.to_numeric(chuva['Abril'].str.replace('---', '93.64'))

chuva['Maio'] = pd.to_numeric(chuva['Maio'].str.replace('---', '86.67'))

chuva['Junho'] = pd.to_numeric(chuva['Junho'].str.replace('---', '59.66'))

chuva['Julho'] = pd.to_numeric(chuva['Julho'].str.replace('---', '54.6'))

chuva['Agosto'] = pd.to_numeric(chuva['Agosto'].str.replace('---', '27.85'))

chuva['Setembro'] = pd.to_numeric(chuva['Setembro'].str.replace('---', '83.8'))

chuva['Outubro'] = pd.to_numeric(chuva['Outubro'].str.replace('---', '139.89'))

chuva['Novembro'] = pd.to_numeric(chuva['Novembro'].str.replace('---', '145.04'))

chuva['Dezembro'] = pd.to_numeric(chuva['Dezembro'].str.replace('---', '236.04'))
```

```python
chuva
```

```python
chuva.dtypes
```

```python
# excluindo as duas últimas linhas

chuva = chuva.drop([36,37])
```

```python
chuva.to_csv('Data/chuva_tratado.csv', index=False)
```

# 2.0 Criação da Série Temporal

```python
chuva2 = chuva.drop(columns='Ano')
chuva2.head()
```

```python
# transformando em array

chuva2 = chuva2.values
```

```python
# transformar em lista

chuva2 = list(chuva2.flatten())
chuva2
```

```python
# criando o índice

indice = pd.date_range('1985', periods=len(chuva2), freq='M')
```

```python
# criando a série

serie = pd.Series(chuva2, index=indice)
serie
```

```python
plt.figure(figsize=(18,8))
serie.plot();
```

# 3.0 Análise e Preparação da Série

## 3.1 Média Móvel

```python
media_movel = serie.rolling(window=6)
media_movel = media_movel.mean()
```

```python
plt.plot(media_movel);
```

```python
plt.plot(serie, label='Série Real')
plt.plot(media_movel, color='red', label='Média Móvel 6 Meses' )
plt.legend(loc='best')
plt.title('PRECIPITAÇÃO NA CIDADE DE SÃO PAULO', fontweight='bold')
plt.show();
```

## 3.2 Decomposição

```python
decomposicao = seasonal_decompose(serie)
decomposicao.plot();
```

## 3.3 Normalidade e Transformação

### 3.3.1 QQ Plot

```python
stats.probplot(serie, dist='norm', plot=plt)
plt.title('NORMAL QQ PLOT', fontweight='bold')
plt.show()
```

### 3.3.2 Shapiro-Wilk

Nível de significância de 5%

- Ho = distribuição normal: p > 0,05
- Ha = distribuição não normal : p <= 0,05

```python
e, p = stats.shapiro(serie)

print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

### 3.3.3 Transformação da série

Diminuir a variância e melhorar a normalidade

```python
# transformação por log

serie2 = np.log(serie)
```

```python
e, p = stats.shapiro(serie2)

print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

```python
# transformação por raiz cúbica (quando possui dados com valor zero ou negativos)

serie3 = (serie)**(1/3)
```

```python
stats.probplot(serie3, dist='norm', plot=plt)
plt.title('NORMAL QQ PLOT', fontweight='bold')
plt.show()
```

```python
e, p = stats.shapiro(serie3)

print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

```python

plt.subplot(2,1,1)
sns.distplot(serie)

plt.subplot(2,1,2)
sns.distplot(serie3)
```

## 3.4 Estacionaridade

Teste KPSS

- Ho = não é estacionário: estatística do teste > valor crítico
- Ha = é estacionário: estatística do teste < valor crítico

```python
kpss = statsmodels.tsa.stattools.kpss(serie3)
print('Estatística do teste: {:.4f}'.format(kpss[0]))
print('p-valor: {:.4f}'.format(kpss[1]))
print('Valores críticos:')

for chave, valor in kpss[3].items():
    print('{}: {:.4f}'.format(chave, valor));
```

A estatística do teste indicou uma série estacionária.

## 3.5 Autocorrelação

```python
plot_acf(serie3, lags=20);
```

Os valores da série estão autocorrelacionados.

```python
plot_pacf(serie3, lags=20);
```

A série é autocorrelacionada, e esse é um pressuposto da análise dos resíduos dos modelos de séries temporais.

# 4.0 Modelo AR

- Modelo arima: (p, d, q)
- Modelo ar: (p, 0, 0)

p: ordem autorregressiva, d: grau de diferenciação, q: ordem da média móvel

```python
modelo_ar = ARIMA(serie3, order=(9, 0, 0));
```

```python
resultado = modelo_ar.fit()
print(resultado.summary())
```

Melhor AIC: 1337.922 = (9, 0, 0)

## 4.1 Análise dos Resíduos

```python
residuos = resultado.resid
```

```python
residuos.plot();
```

A média dos resíduos parece tender a zero.

### 4.1.1 Normalidade dos resíduos

```python
# qq plot

stats.probplot(residuos, dist='norm', plot=plt)
plt.title('NORMAL QQ PLOT', fontweight='bold')
plt.show()
```

```python
# shapiro-wilk

e, p = stats.shapiro(residuos)

print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

```python
sns.distplot(residuos);
```

### 4.1.2 Autocorrelação dos resíduos

```python
plot_acf(residuos, lags=20);
```

```python
plot_pacf(residuos, lags=20);
```

Não existe autocorrelação nos valores dos resíduos.

```python
plt.plot(serie3, label='Série Real')
plt.plot(serie3-residuos, color='red', label='Resíduos')
plt.legend(loc='best')
plt.show();
```

## 4.2 Previsão

```python
resultado.fittedvalues
```

```python
previsao = resultado.predict(start=431, end=443)
previsao
```

```python
previsao2 = resultado.forecast(12)
previsao2[0]
```

```python
plt.plot(serie3, label='Série Real')
plt.plot(previsao, color='purple', label='Previsao')
plt.plot(serie3-residuos, color='orange', label='Resíduos')
plt.legend(loc='best')
plt.show();
```

```python
prev_escala = pd.DataFrame(previsao ** 3, columns=['Previsao_AR'])
prev_escala
```

```python
prev_escala.plot();
```

# 8.0 Modelo AUTO ARIMA

- Modelo Arima:(p,d,q)

- Modelo SARIMA: (p,d,q)(P,D,Q)

```python
modelo_auto = auto_arima(serie3, trace = True, stepwise = False, seasonal=True, max_p=10, max_q=10,
                          max_P=4, max_Q=4, start_p=0, start_q=0, start_P=0, start_Q=0, m=12)

# Trace: apresenta no console a lista dos modelos.
# stepwise: seleção gradual(processo mais rápido, porém menos minucioso).
# m: período sazonal (neste caso mensal).
# Interessante dobrar as ordens máximas: max.p = 10, max.q = 10, max.P = 4, max.Q = 4
```

```python
print(modelo_auto.aic())
```

```python
resultado_auto = modelo_auto.fit(serie3)
print(resultado_auto.summary())
```

## 9.1 Desempenho dos modelos

```python
# Lista com valores reais de 2021
lista = [373.3, 174.1, 137.8, 55.7]
valores_reais = pd.DataFrame(lista, columns = ['valores reais'])
print(valores_reais)
```

```python
auto = prev_escala_auto.iloc[0:4]
auto
```

```python
desempenho = pd.concat([valores_reais, auto],axis=1)
desempenho
```

```python
ar = prev_escala.iloc[1:5]
ar
```

```python
pd.DataFrame.reset_index(ar, drop=True, inplace=True)
ar
```

```python
ma = prev_escala_ma.iloc[1:5]
ma
```

```python
pd.DataFrame.reset_index(ma, drop=True, inplace=True)
ma
```

```python
arma = prev_escala_arma.iloc[1:5]
arma
```

```python
pd.DataFrame.reset_index(arma, drop=True, inplace=True)
arma
```

```python
arima = prev_escala_arima.iloc[0:4]
arima
```

```python
desempenho = pd.concat([desempenho, ar, ma, arma, arima],axis=1)
desempenho
```

### 9.1.1 Erro médio absoluto (MAE)

```python
print('ERRO MÉDIO ABSOLUTO (MAE)')
mae_sarima = mean_absolute_error(desempenho['valores reais'], desempenho['Previsão_SARIMA'])
print('SARIMA: {:.2f}'.format(mae_sarima))
mae_ar = mean_absolute_error(desempenho['valores reais'], desempenho['Previsao_AR'])
print('AR: {:.2f}'.format(mae_ar))
mae_ma = mean_absolute_error(desempenho['valores reais'], desempenho['Previsao_MA'])
print('MA: {:.2f}'.format(mae_ma))
mae_arma = mean_absolute_error(desempenho['valores reais'], desempenho['Previsão_ARMA'])
print('ARMA: {:.2f}'.format(mae_arma))
mae_arima = mean_absolute_error(desempenho['valores reais'], desempenho['Previsão_ARIMA'])
print('ARIMA: {:.2f}'.format(mae_arima))
```

### 9.1.2 Erro quadrático médio (MSE)

```python
print('ERRO QUADRÁTICO MÉDIO (MSE)')
mse_sarima = mean_squared_error(desempenho['valores reais'], desempenho['Previsão_SARIMA'])
print('SARIMA: {:.2f}'.format(mse_sarima))
mse_ar = mean_squared_error(desempenho['valores reais'], desempenho['Previsao_AR'])
print('AR: {:.2f}'.format(mse_ar))
mse_ma = mean_squared_error(desempenho['valores reais'], desempenho['Previsao_MA'])
print('MA: {:.2f}'.format(mse_ma))
mse_arma = mean_squared_error(desempenho['valores reais'], desempenho['Previsão_ARMA'])
print('ARMA: {:.2f}'.format(mse_arma))
mse_arima = mean_squared_error(desempenho['valores reais'], desempenho['Previsão_ARIMA'])
print('ARIMA: {:.2f}'.format(mse_arima))
```

### 9.1.3 Raiz do erro quadrático médio (RMSE)

```python
print('RAIZ DO ERRO QUADRÁTICO MÉDIO (RMSE)')
rmse_sarima = mean_squared_error(desempenho['valores reais'], desempenho['Previsão_SARIMA'], squared=False)
print('SARIMA: {:.2f}'.format(rmse_sarima))
rmse_ar = mean_squared_error(desempenho['valores reais'], desempenho['Previsao_AR'], squared=False)
print('AR: {:.2f}'.format(rmse_ar))
rmse_ma = mean_squared_error(desempenho['valores reais'], desempenho['Previsao_MA'], squared=False)
print('MA: {:.2f}'.format(rmse_ma))
rmse_arma = mean_squared_error(desempenho['valores reais'], desempenho['Previsão_ARMA'], squared=False)
print('ARMA: {:.2f}'.format(rmse_arma))
rmse_arima = mean_squared_error(desempenho['valores reais'], desempenho['Previsão_ARIMA'], squared=False)
print('ARIMA: {:.2f}'.format(rmse_arima))
```
