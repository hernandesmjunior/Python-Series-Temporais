# 0.0 Importações

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import statsmodels.tsa.stattools

from matplotlib.pylab import rcParams

sns.set()
```

```python
rcParams['figure.figsize'] = 18,8
```

# 1.0 Transformação

Tentar transformar a série em aproximadamente normal.

**AirPassengers**: dados clássicos da companhia aérea Box & Jenkins. Totais mensais de passageiros de companhias aéreas internacionais, 1949 a 1960.

```python
serie = pd.read_csv('Data/AirPassengers.csv', sep=',')
```

```python
serie
```

```python
serie = pd.Series(serie['#Passengers'].values, index=serie['Month'])
serie
```

```python
serie.plot()
```

## 1.1 Teste de Normalidade

### 1.1.1 Normal QQ Plot

```python
stats.probplot(serie, dist='norm', plot=plt)
plt.title('NORMAL QQ PLOT', fontweight='bold')
plt.show()
```

### 1.1.2 Shapiro-Wilk

Nível de significância de 0,05 ou 5%

Quando p>0,05 (distribuição normal)

```python
e, p = stats.shapiro(serie)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

## 1.2 Transformação por log (diminuir variância e melhorar a normalidade)

```python
serie2 = np.log(serie)
serie2
```

```python
# normal qq plot

stats.probplot(serie2, dist='norm', plot=plt)
plt.title('NORMAL QQ PLOT', fontweight='bold')
plt.show()
```

```python
# shapiro-wilk

e, p = stats.shapiro(serie2)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

## 1.3 Transformação por raiz cúbica (quando possui dados com valor zero ou negativos)

```python
serie3 = (serie)**(1/3)
serie3
```

```python
# normal qq plot

stats.probplot(serie3, dist='norm', plot=plt)
plt.title('NORMAL QQ PLOT', fontweight='bold')
plt.show()
```

```python
# shapiro-wilk

e, p = stats.shapiro(serie3)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

# 2.0 Diferenciação

Transformar uma série não estacionária em estacionária.

## 2.1 Testes de Estacionaridade

### 2.1.1 Teste KPSS

- Ho = não é estacionário: estatística do teste > valor crítico;
- Ha = é estacionário: estatística do teste < valor crítico.

```python
kpss = statsmodels.tsa.stattools.kpss(serie3)

print('Estatística do teste: {:.4f}'.format(kpss[0]))
print('p-valor: {:.4f}'.format(kpss[1]))
print('Valores críticos:')

for chave, valor in kpss[3].items():
    print('{}: {:.4f}'.format(chave, valor));
```

### 2.1.2 Teste DF

- Ho = não é estacionário: estatística do teste > valor crítico;
- Ha = é estacionário: estatística do teste < valor crítico.

```python
df = statsmodels.tsa.stattools.adfuller(serie3)

print('Estatística do teste: {:.4f}'.format(df[0]))
print('p-valor: {:.8f}'.format(df[1]))
print('Valores críticos:')

for chave, valor in df[4].items():
    print('{}: {:.4f}'.format(chave, valor));
```
