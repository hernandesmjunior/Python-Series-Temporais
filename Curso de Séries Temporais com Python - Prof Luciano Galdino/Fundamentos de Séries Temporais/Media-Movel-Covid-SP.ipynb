# 0.0 Importações

## 0.1 Bibliotecas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from matplotlib.pylab import rcParams

sns.set()
```

```python
rcParams['figure.figsize'] = 18,8
```

## 0.2 Dataset

```python
covid = pd.read_csv('Data/dados_covid_sp.csv', sep=';')
```

```python
covid.head()
```

```python
covid.shape
```

# 1.0 Análise e tratamento inicial

```python
covid.dtypes
```

```python
# substituir vírgula por ponto

covid['casos_mm7d'] = covid['casos_mm7d'].str.replace(',','.')
covid['obitos_mm7d'] = covid['obitos_mm7d'].str.replace(',','.')

# transformar em float

covid['casos_mm7d'] = covid['casos_mm7d'].astype(float)
covid['obitos_mm7d'] = covid['obitos_mm7d'].astype(float)

```

```python
# valores faltantes

covid.isnull().sum()
```

```python
covid_sp = covid[covid['nome_munic']=='São Paulo']
covid_sp.head()
```

# 2.0 Média Móvel de Casos Novos

```python
# média móvel do dataframe

plt.plot(covid_sp['casos_mm7d'])
plt.title('MÉDIA MOVEL DE CASOS NOVOS')
plt.show()
```

```python
# criando a nova serie

serie = pd.Series(covid_sp['casos_novos'].values, index=covid_sp['datahora'])
serie
```

```python
serie.plot();
```

```python
# criando a média movel

media_casos = serie.rolling(window=7)
media_casos = media_casos.mean()

```

```python
plt.figure(figsize=(60,20))
plt.xticks(rotation=90)
plt.plot(media_casos)
```

```python
plt.figure(figsize=(60,20))
plt.xticks(rotation=90)
plt.title('MÉDIA MÓVEL DOS NOVOS CASOS DE COVID-19 DA CIDADE DE SÃO PAULO',fontsize=18, fontweight='bold')
plt.plot(serie, label='Série Real')
plt.plot(media_casos, color='red', label='Média Móvel 7 dias')
plt.legend(loc='upper left')
plt.show()
```

# 3.0 Média Móvel de Óbitos Novos

```python
# média móvel do dataframe

plt.plot(covid_sp['obitos_mm7d'])
plt.title('MÉDIA MOVEL DE ÓBITOS NOVOS')
plt.show()
```
