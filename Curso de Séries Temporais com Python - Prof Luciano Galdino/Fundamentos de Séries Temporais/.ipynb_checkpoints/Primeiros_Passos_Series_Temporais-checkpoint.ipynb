# CRIAÇÃO DE UMA SÉRIE TEMPORAL ALEATÓRIA COM DISTRIBUIÇÃO NORMAL

# 0.0 Importações

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
import datetime

from matplotlib.pylab import rcParams

sns.set()
```

```python
rcParams['figure.figsize'] = 15, 6
```

# 1.0 Série Anual

```python
# série de 1980 a 2020

np.random.seed(10)

dados1 = np.random.normal(0, 1, 41)

dados1
```

```python
type(dados1)
```

```python
# transformando em séries

serie = pd.Series(dados1)
```

```python
type(serie)
```

```python
serie.plot()
```

```python
dados1 = pd.DataFrame(dados1)
dados1.head()
```

```python
dados1.columns = ['valores']
```

```python
dados1.shape
```

```python
dados1.describe()
```

```python
indice = pd.date_range('1980', periods=len(dados1), freq='Y')
indice
```

```python
serie1 = pd.Series(dados1['valores'].values, index=indice)
```

```python
type(serie1)
```

```python
serie1.plot()
```

## 1.1 Verificando a normalidade

### 1.1.1 Normal QQ Plot

```python
stats.probplot(serie1, dist="norm", plot=plt)
plt.title('NORMAL QQ PLOT', fontweight="bold", fontsize=16)
plt.show()
```

### 1.1.2 Teste de Shapiro-Wilk

Quando p> 0,05 (distribuição normal)

```python
e, p = stats.shapiro(serie1)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```

# 2.0 Série Mensal

```python
np.random.seed(6)
dados2 = np.random.normal(0, 1, 72)
dados2
```

```python
dados2 = pd.DataFrame(dados2)
dados2.columns = ['valores']
```

```python
data = np.array('2015-01', dtype=np.datetime64())
data
```

```python
data = data + np.arange(72)
data
```

```python
data = pd.DataFrame(data)
```

```python
data.columns = ['data']
data.head()
```

```python
serie2 = pd.concat([data, dados2], axis=1)
serie2.head()
```

```python
serie2 = pd.Series(serie2['valores'].values, index=serie2['data'])
serie2
```

```python
serie2.plot()
```

## 2.1 Verificando a normalidade

### 2.1.1 Normal QQ Plot

```python
stats.probplot(serie2, dist='norm', plot=plt)
plt.title('NORMAL QQ PLOT', fontweight="bold", fontsize=16)
plt.show()
```

### 2.1.2 Teste de Shapiro-Wilk

```python
e,p = stats.shapiro(serie2)
print('Estatística de teste: {}'.format(e))
print('p-valor: {}'.format(p))
```
